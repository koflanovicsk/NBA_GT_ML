```python
#delete the pos_x column and rename the pos_y column to pos
df = df.drop(columns=['Pos_x'])
df = df.rename(columns={'Pos_y': 'Pos'})
#same for the age columns
df = df.drop(columns=['Age_x'])
df = df.rename(columns={'Age_y': 'Age'})
#same for all the other columns where there are duplicates
df = df.drop(columns=['G_x'])
df = df.rename(columns={'G_y': 'G'})
df = df.drop(columns=['MP_x'])
df = df.rename(columns={'MP_y': 'MP'})

#drop the rows where the salary is NaN
#df = df.dropna(subset=['Salary'])
```
